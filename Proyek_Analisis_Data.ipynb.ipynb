### Gathering Data

```python
customers_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/customers_dataset.csv') 
customers_df.head()
```

```python
geolocation_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/geolocation_dataset.csv')
geolocation_df.head()
```

```python
order_items_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/order_items_dataset.csv')
order_items_df.head()
```

```python
orders_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/orders_dataset.csv')
orders_df.head()
```

```python
products_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/products_dataset.csv')
products_df.head()
```

```python
product_category_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/product_category_name_translation.csv')
product_category_df.head()
```

```python
order_reviews_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/order_reviews_dataset.csv')
order_reviews_df.head()
```

```python
order_pay_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/order_payments_dataset.csv')
order_pay_df.head()
```

```python
sellers_df = pd.read_csv('https://raw.githubusercontent.com/mhmmadgiatt/Proyek-Analisis-Data-E-Commerce/main/data/sellers_dataset.csv')
sellers_df.head()
```

### Assessing Data

1. Memahami setiap informasi dari dataset

```python
print('\n', customers_df.info())
print('\n', geolocation_df.info())
print('\n', order_items_df.info())
print('\n', orders_df.info())
print('\n', products_df.info())
print('\n', product_category_df.info())
print('\n', order_reviews_df.info())
print('\n', order_pay_df.info())
print('\n', sellers_df.info())
```

2. Mengecek apakah ada Missing values pada dataset

```python
print('\nInfo missing values data customers:\n', customers_df.isnull().sum())
print('\nInfo missing values data geolocation:\n', geolocation_df.isnull().sum())
print('\nInfo missing values data order_items:\n', order_items_df.isnull().sum())
print('\nInfo missing values data orders:\n', orders_df.isnull().sum())
print('\nInfo missing values data products:\n', products_df.isnull().sum())
print('\nInfo missing values data product_category:\n', product_category_df.isnull().sum())
print('\nInfo missing values data order_reviews:\n', order_reviews_df.isnull().sum())
print('\nInfo missing values data order_pay:\n', order_pay_df.isnull().sum())
print('\nInfo missing values data sellers:\n', sellers_df.isnull().sum())
```

2. Mendeskripsikan dataset

```python
print('deskripsikan data customers:\n', customers_df.describe())
print('deskripsikan data geolocation:\n', geolocation_df.describe())
print('deskripsikan data order_items:\n', order_items_df.describe())
print('deskripsikan data orders:\n', orders_df.describe())
print('deskripsikan data products:\n', products_df.describe())
print('deskripsikan data product_category:\n', product_category_df.describe())
print('deskripsikan data order_reviews:\n', order_reviews_df.describe())
print('deskripsikan data order_pay:\n', order_pay_df.describe())
print('deskripsikan data sellers:\n', sellers_df.describe())
```

3. Mengecek apakah data Duplikat

```python
print('data duplicate customers:', customers_df.duplicated().sum())
print('data duplicate geolocation:', geolocation_df.duplicated().sum())
print('data duplicate order_items:', order_items_df.duplicated().sum())
print('data duplicate orders:', orders_df.duplicated().sum())
print('data duplicate products:', products_df.duplicated().sum())
print('data duplicate product_category:', product_category_df.duplicated().sum())
print('data duplicate order_reviews:', order_reviews_df.duplicated().sum())
print('data duplicate order_pay:', order_pay_df.duplicated().sum())
print('data duplicate sellers:', sellers_df.duplicated().sum())
```

### Cleaning Data

```python
order_reviews_df = order_reviews_df.dropna(subset=['review_comment_title'])
print('Info missing values data order_reviews:\n', order_reviews_df.isnull().sum())
```

```python
merged_order_df = pd.merge(orders_df, order_items_df, on='order_id')
print(merged_order_df.head())
```

```python
merged_product_df= pd.merge(products_df, product_category_df, on='product_category_name')
print(merged_product_df.head())
```

```python
merged_review_df = pd.merge(merged_order_df, order_reviews_df, on='order_id')
print(merged_review_df.head())
```
